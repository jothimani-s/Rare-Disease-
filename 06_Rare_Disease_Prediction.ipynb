# 06 Rare Disease Prediction

```
import torch
from src.predictor import DiseasePredictor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

## Sample embeddings

```
num_samples=256
embedding_dim=64
num_classes=10

X=torch.randn(num_samples, embedding_dim)
y=torch.randint(0,num_classes,(num_samples,))
```

## Initialize Predictor

```
model=DiseasePredictor(input_dim=64,num_classes=num_classes)
optimizer=torch.optim.Adam(model.parameters(),lr=1e-3)
criterion=torch.nn.CrossEntropyLoss()
```

## Training Loop

```
for epoch in range(5):
    optimizer.zero_grad()
    outputs=model(X)
    loss=criterion(outputs,y)
    loss.backward()
    optimizer.step()
    print(f"Epoch {epoch+1}: Loss={loss.item():.4f}")
```

## Evaluation

```
pred=torch.argmax(model(X),dim=1)

acc=accuracy_score(y.numpy(),pred.numpy())
pre=precision_score(y.numpy(),pred.numpy(),average='macro')
rec=recall_score(y.numpy(),pred.numpy(),average='macro')
f1=f1_score(y.numpy(),pred.numpy(),average='macro')

print("Accuracy :",acc)
print("Precision:",pre)
print("Recall   :",rec)
print("F1-score :",f1)
```

## Save Model

```
torch.save(model.state_dict(),'disease_predictor.pt')
print("Model Saved")
```
