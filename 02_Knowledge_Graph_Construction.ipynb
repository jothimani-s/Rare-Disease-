# 02 Knowledge Graph Construction
Build a biomedical knowledge graph from integrated datasets.

```
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from pathlib import Path
from src.graph_builder import BiomedicalKnowledgeGraph
```

## Load integrated dataset

```
DATA=Path('../datasets/integrated_dataset.csv')
if DATA.exists():
    df=pd.read_csv(DATA)
else:
    df=pd.DataFrame({
        'source':['GeneA','GeneB','Disease1'],
        'relation':['associated_with','causes','treated_by'],
        'target':['Disease1','Disease2','DrugX']
    })
df.head()
```

## Construct graph

```
kg=BiomedicalKnowledgeGraph()
kg.add_dataframe(df,'source','relation','target')
stats=kg.statistics()
print(stats)
```

## Visualize graph

```
plt.figure(figsize=(8,6))
pos=nx.spring_layout(kg.graph,seed=42)
nx.draw_networkx(kg.graph,pos,node_color='lightblue',node_size=900,font_size=8,arrows=True)
plt.axis('off')
plt.show()
```

## Export GraphML

```
Path('../results').mkdir(exist_ok=True)
kg.save_graphml('../results/biomedical_kg.graphml')
print('Graph saved.')
```
